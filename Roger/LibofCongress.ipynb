```python
import requests
```

```python
LoC_collections = requests.get("https://www.loc.gov/collections/?fo=json").json()
```

```python
# get all collections from LIB
while True:
    for collection in LoC_collections["results"]:
        print(collection["title"])
    next_page = LoC_collections["pagination"]["next"]
    if next_page is not None:
        LoC_collections = requests.get(next_page).json()
    else:
        break
```
